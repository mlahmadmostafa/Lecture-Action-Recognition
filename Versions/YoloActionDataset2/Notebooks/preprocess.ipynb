```python
import os
import cv2
import pandas as pd
# Get list of images in train directory
train_dir = '../Dataset_2/EduAction-A'
result_dir = '../Dataset_2/EduAction-A_8Frames'

train_folder = "EduAction"
results_folder = "EduAction-A_8Frames"
df_paths = []

for folder in os.listdir(train_dir):
    os.makedirs(f"{result_dir}/{folder}", exist_ok=True)
    for video in os.listdir(os.path.join(train_dir,folder)):
        video_path = os.path.join(train_dir,folder,video)
        saved_video_path = os.path.join(train_folder,folder,video)
        cap = cv2.VideoCapture(video_path)
        
        
        
        frames = []
        fps = cap.get(cv2.CAP_PROP_FPS)
        width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Output codec
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            frames.append(frame)
        cap.release()
        clip_len = 8
        stride = 1
        for i in range(0, len(frames) - clip_len + 1, stride):
            new_video_path = f"{result_dir}/{folder}/{video}_{i}.mp4"
            saved_new_video_path = f"{results_folder}/{folder}/{video.split('.')[0]}_{i}.mp4"
            out = cv2.VideoWriter(new_video_path, fourcc, fps, (width, height))
            for j in range(i, i + clip_len):
                out.write(frames[j])
            out.release()
            # old_video_path, new_video_path, class
            df_paths.append([video_path[3:], new_video_path[3:], folder])
df_paths = pd.DataFrame(df_paths, columns=['old_video_path', 'new_video_path', 'class'])
```

```python
df_paths.to_csv('../Dataset_2/EduAction-A_8Frames.csv')
```
